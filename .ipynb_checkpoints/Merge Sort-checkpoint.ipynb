```python
# Python program for implementation of MergeSort
# https://www.geeksforgeeks.org/merge-sort/?ref=lbp
def mergeSort(arr):
    if len(arr) > 1:
 
         # Finding the mid of the array
        mid = len(arr)//2
 
        # Dividing the array elements
        L = arr[:mid]
 
        # into 2 halves
        R = arr[mid:]
 
        # Sorting the first half
        mergeSort(L)
 
        # Sorting the second half
        mergeSort(R)
 
        i = j = k = 0
 
        # Copy data to temp arrays L[] and R[]
        while i < len(L) and j < len(R):
            if L[i] <= R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1
 
        # Checking if any element was left
        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1
 
        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1

# Code to print the list
 

def printList(arr):
    for i in range(len(arr)):
        print(arr[i], end=" ")
    print()
 
 
# Driver Code
if __name__ == '__main__':
    arr = [12, 11, 13, 5, 6, 7]
    print("Given array is", end="\n")
    printList(arr)
    mergeSort(arr)
    print("Sorted array is: ", end="\n")
    printList(arr)
 
```

```python
def MergeSort(arr, start, end):
    if start<end:
        middle = (start+end)//2
        #middle = start + (end - start)//2
        MergeSort(arr, start, middle)
        MergeSort(arr, middle+1, end)
        Merge(arr, start, middle, end)
        #printArray(arr)
        
def Merge(arr, start, middle, end):
    lenL = middle-start +1
    lenR = end - middle
    
    leftArr = [None] * lenL
    rightArr = [None] * lenR
    
    for i in range(lenL):
        leftArr[i] = arr[start+i]
    for j in range(lenR):
        rightArr[j] = arr[middle+1+j]
    
    i = 0
    j = 0
    k = start
    while i<lenL and j<lenR:
        if leftArr[i] < rightArr[j]:
            arr[k] = leftArr[i]
            i+=1
            k+=1
        else:
            arr[k] = rightArr[j]
            j+=1
            k+=1
    
    while i<lenL:
        arr[k] = leftArr[i]
        k+=1
        i+=1
    while j<lenR:
        arr[k] = rightArr[j]
        k+=1
        j+=1

def printArr(arr):
    for i in range(len(arr)):
        print('%d' %(arr[i]), end = ' ')
    print('\n')
            

arr1 = [6,9,5,3,2,1,7]
arr2 = [6,9,5,3,2,1,7,11]
MergeSort(arr1, 0, len(arr1)-1)
MergeSort(arr2, 0, len(arr2)-1)
printArr(arr1)
printArr(arr2)
```
